#### Plot Log Data Sumur B

```python
facies_colors = ['#F4D03F', '#F5B041','#DC7633','#6E2C00',
       '#1B4F72','#2E86C1','#7FFF00']
```

```python
def make_facies_log_plot(logs, facies_colors):
    #make sure logs are sorted by DEPTH_MD
    logs = logs.sort_values(by='DEPTH_MD')
    cmap_facies = colors.ListedColormap(
            facies_colors[0:len(facies_colors)], 'indexed')

    ztop=logs.DEPTH_MD.min(); zbot=logs.DEPTH_MD.max()

    cluster=np.repeat(np.expand_dims(logs['FACIES'].values,1), 100, 1)

    f, ax = plt.subplots(nrows=1, ncols=10, figsize=(18, 10))
    ax[0].plot(logs.CALI, logs.DEPTH_MD, '-g')
    ax[1].plot(logs.RMED, logs.DEPTH_MD, '-')
    ax[2].plot(logs.RDEP, logs.DEPTH_MD, '-', color='0.5')
    ax[3].plot(logs.RHOB, logs.DEPTH_MD, '-', color='r')
    ax[4].plot(logs.GR, logs.DEPTH_MD, '-', color='coral')
    ax[5].plot(logs.NPHI, logs.DEPTH_MD, '-', color='black')
    ax[6].plot(logs.PEF, logs.DEPTH_MD, '-', color='navy')
    ax[7].plot(logs.DTC, logs.DEPTH_MD, '-', color='purple')
    ax[8].plot(logs.DRHO, logs.DEPTH_MD, '-', color='violet')

    im=ax[9].imshow(cluster, interpolation='none', aspect='auto',
                    cmap=cmap_facies,vmin=1,vmax=6)

    # divider = make_axes_locatable(ax[9])
    # cax = divider.append_axes("right", size="7%", pad=0.05)
    # cbar=plt.colorbar(im, cax=cax)
    # cbar.set_label((17*' ').join(['1', '2', '3',
    #                             '4', '5', '6']))
    # cbar.set_ticks(range(0,1)); cbar.set_ticklabels('')

    for i in range(len(ax)-1):
        ax[i].set_ylim(ztop,zbot)
        ax[i].invert_yaxis()
        ax[i].grid()
        ax[i].locator_params(axis='x', nbins=3)

    ax[0].set_xlabel("CALI")
    ax[0].set_xlim(logs.CALI.min(),logs.CALI.max())
    ax[1].set_xlabel("RMED")
    ax[1].set_xlim(logs.RMED.min(),logs.RMED.max())
    ax[2].set_xlabel("RDEP")
    ax[2].set_xlim(logs.RDEP.min(),logs.RDEP.max())
    ax[3].set_xlabel("RHOB")
    ax[3].set_xlim(logs.RHOB.min(),logs.RHOB.max())
    ax[4].set_xlabel("GR")
    ax[4].set_xlim(logs.GR.min(),logs.GR.max())
    ax[5].set_xlabel("NPHI")
    ax[5].set_xlim(logs.NPHI.min(),logs.NPHI.max())
    ax[6].set_xlabel("PEF")
    ax[6].set_xlim(logs.PEF.min(),logs.PEF.max())
    ax[7].set_xlabel("DTC")
    ax[7].set_xlim(logs.DTC.min(),logs.DTC.max())
    ax[8].set_xlabel("DRHO")
    ax[8].set_xlim(logs.DRHO.min(),logs.DRHO.max())
    ax[9].set_xlabel("FACIES")

    ax[1].set_yticklabels([]); ax[2].set_yticklabels([]); ax[3].set_yticklabels([])
    ax[4].set_yticklabels([]); ax[9].set_yticklabels([])
    ax[9].set_xticklabels([])
```

```python
make_facies_log_plot(datasetUP,facies_colors)
```

```python
#Split data feature dan facies
X = datasetUP.drop(['DEPTH_MD',"FACIES"], axis=1)
y = datasetUP['FACIES']
```

```python
X.columns
```

```python
#Splitting data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Model Machine

#### K-nearest Neighbors

```python
#Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param_distributions = {'n_neighbors': np.arange(3, 15, 3),
                       'weights': ['uniform', 'distance']}
grid_search = GridSearchCV(KNeighborsClassifier(),
                           param_distributions,
                           scoring='accuracy', cv=10)
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
pd.DataFrame(grid_search.cv_results_).transpose()
```

##### Model Evaluation on Test Set

```python
knn = grid_search.best_estimator_
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_).plot()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

#### Support Vector Machine (SVM)

```python
#Hyperparameter Tuning
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_distributions = {'C': [1, 5, 10],
                       'kernel': ('linear', 'rbf'),
                       'gamma' : ('scale', 'auto')}

grid_search = GridSearchCV(SVC(),
                           param_grid=param_distributions,
                           scoring='accuracy',
                           cv=10)
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
pd.DataFrame(grid_search.cv_results_).transpose()
```

##### Model Evaluation on Test Set

```python
model = grid_search.best_estimator_
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

### Clustering (K-means)

```python
from sklearn.cluster import KMeans
datasetUP.drop(['FACIES'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=scaler.fit_transform(datasetUP[["GR","RHOB","NPHI","DTC"]])

df_scaled
```

```python
def optimise_k_means(data,max_k):
    means = []
    inertias = []

    for k in range(1,max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)

        means.append(k)
        inertias.append(kmeans.inertia_)

    fig = plt.subplots(figsize=(10,5))
    plt.plot(means,inertias, "o-")
    plt.xlabel("Number of CLuster")
    plt.ylabel("Inertia")
    plt.grid(True)
    plt.show()
```

```python
optimise_k_means(datasetUP[["GR","RHOB","NPHI","DTC"]], 30)
```

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(datasetUP[["GR","RHOB","NPHI","DTC"]])
```

```python
datasetUP['CLUSTERS'] = kmeans.labels_
```

```python
cluster_colors = ['#F4D03F', '#F5B041','#DC7633','#6E2C00',
       '#1B4F72','#2E86C1','#7FFF00','#9932CC']
```

```python
def make_cluster_log_plot(logs, cluster_colors):
    #make sure logs are sorted by DEPTH_MD
    logs = logs.sort_values(by='DEPTH_MD')
    cmap_cluster = colors.ListedColormap(
            cluster_colors[0:len(cluster_colors)], 'indexed')

    ztop=logs.DEPTH_MD.min(); zbot=logs.DEPTH_MD.max()

    cluster=np.repeat(np.expand_dims(logs['CLUSTERS'].values,1), 100, 1)

    f, ax = plt.subplots(nrows=1, ncols=10, figsize=(18, 10))
    ax[0].plot(logs.CALI, logs.DEPTH_MD, '-g')
    ax[1].plot(logs.RMED, logs.DEPTH_MD, '-')
    ax[2].plot(logs.RDEP, logs.DEPTH_MD, '-', color='0.5')
    ax[3].plot(logs.RHOB, logs.DEPTH_MD, '-', color='r')
    ax[4].plot(logs.GR, logs.DEPTH_MD, '-', color='coral')
    ax[5].plot(logs.NPHI, logs.DEPTH_MD, '-', color='black')
    ax[6].plot(logs.PEF, logs.DEPTH_MD, '-', color='navy')
    ax[7].plot(logs.DTC, logs.DEPTH_MD, '-', color='purple')
    ax[8].plot(logs.DRHO, logs.DEPTH_MD, '-', color='violet')

    im=ax[9].imshow(cluster, interpolation='none', aspect='auto',
                    cmap=cmap_cluster,vmin=1,vmax=5)

    # divider = make_axes_locatable(ax[9])
    # cax = divider.append_axes("right", size="7%", pad=0.05)
    # cbar=plt.colorbar(im, cax=cax)
    # cbar.set_label((17*' ').join(['1', '2', '3',
    #                             '4', '5', '6']))
    # cbar.set_ticks(range(0,1)); cbar.set_ticklabels('')

    for i in range(len(ax)-1):
        ax[i].set_ylim(ztop,zbot)
        ax[i].invert_yaxis()
        ax[i].grid()
        ax[i].locator_params(axis='x', nbins=3)

    ax[0].set_xlabel("CALI")
    ax[0].set_xlim(logs.CALI.min(),logs.CALI.max())
    ax[1].set_xlabel("RMED")
    ax[1].set_xlim(logs.RMED.min(),logs.RMED.max())
    ax[2].set_xlabel("RDEP")
    ax[2].set_xlim(logs.RDEP.min(),logs.RDEP.max())
    ax[3].set_xlabel("RHOB")
    ax[3].set_xlim(logs.RHOB.min(),logs.RHOB.max())
    ax[4].set_xlabel("GR")
    ax[4].set_xlim(logs.GR.min(),logs.GR.max())
    ax[5].set_xlabel("NPHI")
    ax[5].set_xlim(logs.NPHI.min(),logs.NPHI.max())
    ax[6].set_xlabel("PEF")
    ax[6].set_xlim(logs.PEF.min(),logs.PEF.max())
    ax[7].set_xlabel("DTC")
    ax[7].set_xlim(logs.DTC.min(),logs.DTC.max())
    ax[8].set_xlabel("DRHO")
    ax[8].set_xlim(logs.DRHO.min(),logs.DRHO.max())
    ax[9].set_xlabel('KMEANS')

    ax[1].set_yticklabels([]); ax[2].set_yticklabels([]); ax[3].set_yticklabels([])
    ax[4].set_yticklabels([]); ax[9].set_yticklabels([])
    ax[9].set_xticklabels([])
```
